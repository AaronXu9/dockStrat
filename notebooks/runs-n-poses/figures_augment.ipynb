```python
bust_dfs = {}
# Create posebusters_results directory if it doesn't exist
posebusters_dir = data_dir / "posebusters_results"
posebusters_dir.mkdir(exist_ok=True)

for m in plotting.METHODS:
    if m == 'gnina':
        # For GNINA, we can extract PoseBusters-like data from the existing results
        gnina_path = "/home/aoxu/projects/PoseBench/notebooks/05_specialized_analysis/gnina_runsNposes_0_results.csv"
        if os.path.exists(gnina_path):
            print(f"Processing GNINA PoseBusters data from {gnina_path}")
            gnina_full = pd.read_csv(gnina_path)
            
            # Create PoseBusters-compatible dataframe from GNINA results
            gnina_bust = pd.DataFrame()
            gnina_bust['system_id'] = gnina_full['protein']
            gnina_bust['ligand_instance_chain'] = gnina_full['protein'].apply(
                lambda x: x.split('__')[-1] if '__' in x else '1.L'
            )
            
            # Map GNINA columns to PoseBusters columns
            gnina_bust['pb_success'] = (gnina_full['rmsd'] <= 2.0).astype(float)  # Success if RMSD <= 2Å
            gnina_bust['sanitization'] = gnina_full['sanitization']
            gnina_bust['inchi_convertible'] = gnina_full['inchi_convertible'] 
            gnina_bust['all_atoms_connected'] = gnina_full['all_atoms_connected']
            gnina_bust['bond_lengths'] = gnina_full['bond_lengths']
            gnina_bust['bond_angles'] = gnina_full['bond_angles']
            gnina_bust['internal_steric_clash'] = gnina_full['internal_steric_clash']
            gnina_bust['aromatic_ring_flatness'] = gnina_full['aromatic_ring_flatness']
            gnina_bust['non-aromatic_ring_non-flatness'] = True  # Not available in GNINA
            gnina_bust['double_bond_flatness'] = gnina_full['double_bond_flatness']
            gnina_bust['internal_energy'] = gnina_full['internal_energy']
            gnina_bust['protein-ligand_maximum_distance'] = gnina_full['protein-ligand_maximum_distance']
            gnina_bust['minimum_distance_to_protein'] = gnina_full['minimum_distance_to_protein']
            gnina_bust['volume_overlap_with_protein'] = gnina_full['volume_overlap_with_protein']
            
            bust_dfs[m] = gnina_bust
            print(f"Created GNINA PoseBusters data: {len(gnina_bust)} rows")
        else:
            print(f"Warning: GNINA results file not found, creating empty dataframe")
            bust_dfs[m] = pd.DataFrame(columns=[
                "system_id", "ligand_instance_chain", "pb_success",
                "sanitization", "inchi_convertible", "all_atoms_connected",
                "bond_lengths", "bond_angles", "internal_steric_clash",
                "aromatic_ring_flatness", "non-aromatic_ring_non-flatness",
                "double_bond_flatness", "internal_energy",
                "protein-ligand_maximum_distance", "minimum_distance_to_protein",
                "volume_overlap_with_protein"
            ])
    else:
        filename = data_dir / "posebusters_results" / f"{m}.csv"
        if filename.exists():
            bust_dfs[m] = pd.read_csv(filename)
        else:
            # Create empty dataframe with required columns if file doesn't exist
            print(f"Warning: {filename} not found, creating empty dataframe")
            bust_dfs[m] = pd.DataFrame(columns=[
                "system_id", "ligand_instance_chain", "pb_success",
                "sanitization", "inchi_convertible", "all_atoms_connected",
                "bond_lengths", "bond_angles", "internal_steric_clash",
                "aromatic_ring_flatness", "non-aromatic_ring_non-flatness",
                "double_bond_flatness", "internal_energy",
                "protein-ligand_maximum_distance", "minimum_distance_to_protein",
                "volume_overlap_with_protein"
            ])
```

```python
full_datasets = {}

# Process existing methods from prediction files
for method in plotting.METHODS:
    if method == 'gnina':
        continue  # Skip gnina here, we'll handle it separately
        
    filename = data_dir / "predictions" / f"{method}.csv"
    if not filename.exists():
        print(f"Warning: {filename} not found, skipping method {method}")
        continue
        
    try:
        df = pd.read_csv(filename, low_memory=False)
        print(f"Loaded {method}: {len(df)} rows")
    except Exception as e:
        print(f"Error loading {method}: {e}")
        continue
        
    keep_columns = [
        "target",
        "ligand_instance_chain",
        "lddt_pli",
        "rmsd",
        "lddt_lp",
        "bb_rmsd",
        "seed",
        "sample",
        "ranking_score",
        "ligand_is_proper",
        "prot_lig_chain_iptm_average_lddt_pli",
        "prot_lig_chain_iptm_min_lddt_pli",
        "prot_lig_chain_iptm_max_lddt_pli",
        "lig_prot_chain_iptm_average_lddt_pli",
        "lig_prot_chain_iptm_min_lddt_pli",
        "lig_prot_chain_iptm_max_lddt_pli",
        "prot_lig_chain_iptm_average_rmsd",
        "prot_lig_chain_iptm_min_rmsd",
        "prot_lig_chain_iptm_max_rmsd",
        "lig_prot_chain_iptm_average_rmsd",
        "lig_prot_chain_iptm_min_rmsd",
        "lig_prot_chain_iptm_max_rmsd",
        "model_ligand_chain_lddt_pli",
        "model_ligand_chain_rmsd",
        "ligand_ccd_code",
        "model_ligand_smiles",
        "pred_pocket_f1",
    ]
    if "seed" not in df.columns:
        df["seed"] = 1
    if "sample" not in df.columns:
        df["sample"] = 1
    if "ranking_score" not in df.columns:
        df["ranking_score"] = 1
    if "lig_prot_chain_iptm_average_rmsd" not in df.columns:
        df["lig_prot_chain_iptm_average_rmsd"] = 1
    if "prot_lig_chain_iptm_average_rmsd" not in df.columns:
        df["prot_lig_chain_iptm_average_rmsd"] = 1
    if "pred_pocket_f1" not in df.columns:
        df["pred_pocket_f1"] = 1
    keep_columns = [c for c in keep_columns if c in df.columns]
    full_datasets[method] = (
        df[keep_columns].rename(columns={"target": "system_id"}).reset_index(drop=True)
    )
    full_datasets[method]["group_key"] = (
        full_datasets[method]["system_id"]
        + "__"
        + full_datasets[method]["ligand_instance_chain"]
    )
    full_datasets[method]["method"] = method
    full_datasets[method] = (
        full_datasets[method]
        .sort_values(by=["lddt_pli", "rmsd"], ascending=[False, True])
        .groupby(["group_key", "seed", "sample"])
        .head(1)
        .reset_index(drop=True)
    )
    if method in bust_dfs:
        full_datasets[method] = full_datasets[method].merge(
            bust_dfs[method][["system_id", "ligand_instance_chain", "pb_success"]],
            on=["system_id", "ligand_instance_chain"],
            how="left",
        )
        full_datasets[method]["pb_success"] = (
            full_datasets[method]["pb_success"].fillna(False).astype(float)
        )
    else:
        full_datasets[method]["pb_success"] = -1

# Add GNINA data if available
if 'gnina_processed' in locals() and gnina_processed is not None:
    print("Adding GNINA to full_datasets...")
    
    # Select the best pose per system (rank 1)
    gnina_best = gnina_processed[gnina_processed['rank'] == 1].copy()
    
    # Process like other methods
    gnina_best = gnina_best.rename(columns={"target": "system_id"})
    gnina_best["method"] = "gnina"
    
    # Add to full_datasets
    full_datasets['gnina'] = gnina_best
    full_datasets['gnina']["pb_success"] = -1  # No PoseBusters data for GNINA
    
    print(f"Added GNINA with {len(gnina_best)} systems")

print(f"Successfully loaded {len(full_datasets)} methods: {list(full_datasets.keys())}")

# Update plotting methods to reflect actually loaded data
plotting.METHODS = list(full_datasets.keys())
plotting.COMMON_SUBSET_METHODS = [m for m in plotting.COMMON_SUBSET_METHODS if m in full_datasets.keys()]

print(f"Final methods for plotting: {plotting.METHODS}")
print(f"Common subset methods: {plotting.COMMON_SUBSET_METHODS}")
```

```python
# Load and process GNINA results
gnina_results_path = "/home/aoxu/projects/PoseBench/notebooks/05_specialized_analysis/gnina_runsNposes_0_results.csv"

if os.path.exists(gnina_results_path):
    print("Loading GNINA results...")
    gnina_df = pd.read_csv(gnina_results_path)
    print(f"Loaded GNINA results: {len(gnina_df)} rows")
    
    # Convert GNINA format to match expected prediction format
    gnina_processed = pd.DataFrame()
    gnina_processed['target'] = gnina_df['protein']  # Use protein as target/system_id
    gnina_processed['system_id'] = gnina_df['protein']
    gnina_processed['method'] = 'gnina'
    gnina_processed['rmsd'] = gnina_df['rmsd']
    gnina_processed['kabsch_rmsd'] = gnina_df['kabsch_rmsd'] 
    gnina_processed['score'] = gnina_df['score']
    gnina_processed['rank'] = gnina_df['rank']
    
    # Add ligand instance chain (extract from protein name or default)
    # Assuming protein format is like "7gio__1__1.A_1.B__1.L_1.O"
    gnina_processed['ligand_instance_chain'] = gnina_processed['system_id'].apply(
        lambda x: x.split('__')[-1] if '__' in x else '1.L'
    )
    
    # Add required columns with default values
    gnina_processed['seed'] = 1
    gnina_processed['sample'] = 1
    gnina_processed['ranking_score'] = gnina_df['score']  # Use GNINA score
    gnina_processed['lddt_pli'] = 0.0  # GNINA doesn't have LDDT, set to 0
    gnina_processed['lddt_lp'] = 0.0
    gnina_processed['bb_rmsd'] = gnina_processed['rmsd']  # Use RMSD as backbone RMSD
    gnina_processed['ligand_is_proper'] = True  # Assume all are proper ligands
    
    # Set all IPTM-related columns to default values since GNINA doesn't have them
    iptm_cols = [
        'prot_lig_chain_iptm_average_lddt_pli', 'prot_lig_chain_iptm_min_lddt_pli',
        'prot_lig_chain_iptm_max_lddt_pli', 'lig_prot_chain_iptm_average_lddt_pli',
        'lig_prot_chain_iptm_min_lddt_pli', 'lig_prot_chain_iptm_max_lddt_pli',
        'prot_lig_chain_iptm_average_rmsd', 'prot_lig_chain_iptm_min_rmsd',
        'prot_lig_chain_iptm_max_rmsd', 'lig_prot_chain_iptm_average_rmsd',
        'lig_prot_chain_iptm_min_rmsd', 'lig_prot_chain_iptm_max_rmsd'
    ]
    for col in iptm_cols:
        gnina_processed[col] = 1.0
    
    # Add remaining columns
    gnina_processed['model_ligand_chain_lddt_pli'] = 0.0
    gnina_processed['model_ligand_chain_rmsd'] = gnina_processed['rmsd']
    gnina_processed['ligand_ccd_code'] = 'UNK'
    gnina_processed['model_ligand_smiles'] = ''
    gnina_processed['pred_pocket_f1'] = 1.0
    
    # Create group_key for consistency
    gnina_processed['group_key'] = (
        gnina_processed['system_id'] + '__' + gnina_processed['ligand_instance_chain']
    )
    
    print(f"Processed GNINA data: {len(gnina_processed)} rows")
    print(f"Unique systems: {gnina_processed['system_id'].nunique()}")
    
    # Add GNINA to methods list
    if 'gnina' not in plotting.METHODS:
        plotting.METHODS.append('gnina')
        plotting.COMMON_SUBSET_METHODS.append('gnina')
        print("Added GNINA to methods list")
    
    # Add GNINA color to plotting if not exists
    if not hasattr(plotting, 'COLORS'):
        plotting.COLORS = {}
    if 'gnina' not in plotting.COLORS:
        plotting.COLORS['gnina'] = '#FF6B6B'  # Red color for GNINA
    
    print(f"Updated methods: {plotting.METHODS}")
    
else:
    print(f"GNINA results file not found at {gnina_results_path}")
    gnina_processed = None
```

```python
def pivot_df(df, annotated_df):
    df = df.pivot(
        index=[
            "group_key",
            "system_id",
            "ligand_is_proper",
            "ligand_instance_chain",
        ],
        columns="method",
        values=[
            "lddt_pli",
            "rmsd",
            "lddt_lp",
            "bb_rmsd",
            "pb_success",
            "pred_pocket_f1",
        ],
    ).reset_index()
    df.columns = [f"{col[0]}_{col[1]}" if len(col[1]) else col[0] for col in df.columns]
    df = df[df["ligand_is_proper"].fillna(False)].reset_index(drop=True)
    merge_columns = [col for col in annotated_df.columns if col not in df.columns]
    df = df.merge(
        annotated_df[["group_key"] + merge_columns], on="group_key", how="left"
    )
    return df


top_dfs = {}
top_5_dfs = {}
best_5_dfs = {}
best_dfs = {}
worst_dfs = {}
random_dfs = {}
random_5_dfs = {}
rank_by = "lig_prot_chain_iptm_average_rmsd"
for m in full_datasets:
    top_dfs[m] = (
        full_datasets[m]
        .sort_values(by=rank_by, ascending=False)
        .groupby(["system_id", "ligand_instance_chain"])
        .head(1)
    )
    best_dfs[m] = (
        full_datasets[m]
        .sort_values(by="lddt_pli", ascending=False)
        .groupby(["system_id", "ligand_instance_chain"])
        .head(1)
    )
    worst_dfs[m] = (
        full_datasets[m]
        .sort_values(by="lddt_pli", ascending=True)
        .groupby(["system_id", "ligand_instance_chain"])
        .head(1)
    )
    random_dfs[m] = (
        full_datasets[m]
        .sample(frac=1)
        .groupby(["system_id", "ligand_instance_chain"])
        .head(1)
    )
    all_top = (
        full_datasets[m]
        .sort_values(by=rank_by, ascending=False)
        .groupby(["group_key", "seed"])
        .head(1)
    )

    all_top["Rank"] = all_top.groupby("group_key")[rank_by].rank(
        ascending=False, method="first"
    )
    top_5_dfs[m] = []

    for rank in range(1, 6):
        df = all_top[all_top["Rank"] == rank]
        top_5_dfs[m].append(df)

    all_best = (
        full_datasets[m]
        .sort_values(by="lddt_pli", ascending=False)
        .groupby(["group_key", "seed"])
        .head(1)
    )
    all_best["Rank"] = all_best.groupby("group_key")["lddt_pli"].rank(
        ascending=False, method="first"
    )
    best_5_dfs[m] = []
    for rank in range(1, 6):
        df = all_best[all_best["Rank"] == rank]
        best_5_dfs[m].append(df)

    all_random = full_datasets[m].sample(frac=1).groupby(["group_key", "seed"]).head(1)
    all_random["Rank"] = all_random.groupby("group_key")[rank_by].rank(
        ascending=False, method="first"
    )
    random_5_dfs[m] = []
    for rank in range(1, 6):
        df = all_random[all_random["Rank"] == rank]
        random_5_dfs[m].append(df)

results_df_top = pivot_df(pd.concat(top_dfs.values()), annotated_df)
results_df_best = pivot_df(pd.concat(best_dfs.values()), annotated_df)
results_df_worst = pivot_df(pd.concat(worst_dfs.values()), annotated_df)
results_df_random = pivot_df(pd.concat(random_dfs.values()), annotated_df)
results_df_top_5 = []
results_df_best_5 = []
results_df_random_5 = []
for i in range(5):
    results_df_top_5.append(
        pivot_df(pd.concat([top_5_dfs[m][i] for m in top_5_dfs]), annotated_df)
    )
    results_df_best_5.append(
        pivot_df(pd.concat([best_5_dfs[m][i] for m in best_5_dfs]), annotated_df)
    )
    results_df_random_5.append(
        pivot_df(pd.concat([random_5_dfs[m][i] for m in random_5_dfs]), annotated_df)
    )

dfs = {
    "top": results_df_top,
    "best": results_df_best,
    "worst": results_df_worst,
    "random": results_df_random,
}
for i in range(5):
    dfs[f"top_5_{i + 1}"] = results_df_top_5[i]
    dfs[f"best_5_{i + 1}"] = results_df_best_5[i]
    dfs[f"random_5_{i + 1}"] = results_df_random_5[i]

for df_name in dfs:
    dfs[df_name]["lddt_pli_max"] = np.nanmax(
        dfs[df_name][
            [
                f"lddt_pli_{m}"
                for m in plotting.METHODS
                if f"lddt_pli_{m}" in dfs[df_name].columns
            ]
        ],
        axis=1,
    )
    dfs[df_name]["rmsd_min"] = np.nanmin(
        dfs[df_name][
            [
                f"rmsd_{m}"
                for m in plotting.METHODS
                if f"rmsd_{m}" in dfs[df_name].columns
            ]
        ],
        axis=1,
    )
    dfs[df_name]["lddt_pli_average"] = np.nanmedian(
        dfs[df_name][
            [
                f"lddt_pli_{m}"
                for m in plotting.METHODS
                if f"lddt_pli_{m}" in dfs[df_name].columns
            ]
        ],
        axis=1,
    )
    dfs[df_name]["rmsd_average"] = np.nanmedian(
        dfs[df_name][
            [
                f"rmsd_{m}"
                for m in plotting.METHODS
                if f"rmsd_{m}" in dfs[df_name].columns
            ]
        ],
        axis=1,
    )

common_subset_dfs_all = {}
cluster_dfs_all = {}
for df in dfs:
    common_subset_dfs_all[df] = (
        dfs[df]
        .dropna(
            subset=[f"lddt_pli_{method}" for method in plotting.COMMON_SUBSET_METHODS]
            + ["sucos_shape"]
        )
        .reset_index(drop=True)
    )
    cluster_dfs_all[df] = (
        common_subset_dfs_all[df][
            common_subset_dfs_all[df]["ligand_is_proper"]
            & (common_subset_dfs_all[df]["sucos_shape"].notna())
        ]
        .sort_values(by=plotting.SIMILARITY_METRIC)
        .groupby("cluster")
        .head(1)
    )
```
